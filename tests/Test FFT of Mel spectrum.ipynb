## Prepare the environment

```python
import sys
sys.path = ['..'] + sys.path # this is where the features module is stored
import scipy
import matplotlib

%pylab inline
figsize(10,4)
matplotlib.rcParams.update({'figure.dpi' : 300,
                            'lines.linewidth' : 1})


def stft(x, fs, framesz, hop):
    framesamp = int(framesz*fs)
    hopsamp = int(hop*fs)
    w = scipy.hamming(framesamp)
    X = scipy.array([scipy.fft(w*x[i:i+framesamp]) 
                     for i in range(0, len(x)-framesamp, hopsamp)])
    return X

def half2Darray(arr):
    return arr[:,:arr.shape[1]/2]
```

## Prepare the data

```python
fs   = 44100.
secs = 10.
freq = 8000.
t    = linspace(0, secs, secs*fs)
```

Our signal $y$ is a combination of a sine wave $y_1$ and no signal (0 values) at a frequency determined by `INTERLEAVE_FREQ`. This is done by generating another sinewave, $y_2$, with frequency `INTERLEAVE_FREQ`, a setting the signal $y$ with value 0 where $y_2 \leq 0$ and with $y_1$ otherwise. 

```python
''' legacy stuff '''
#y2 = 0.2*sin(4 * pi * freq * t)
#y2 = [0] * int(secs*fs)
#sl = arange(0, fs*secs, fs*secs/4.).astype(int)
#y  = hstack( (y1[sl[0]:sl[1]], y2[sl[1]:sl[2]], y1[sl[2]:sl[3]], y2[sl[3]:None]) )

INTERLEAVE_FREQ = 6 # 6 Hz, i.e. 6 times a seconds.

y1 = sin(2 * pi * freq * t)
y2 = sin(2 * pi * INTERLEAVE_FREQ * t)

y = zeros ( (len(y1), ))
for i in range(len(y1)):
    y[i] = y1[i] if y2[i] > 0 else 0


plot(t, y1, "--", label="y1", color="blue")
plot(t, y,        label="y", color="red", linewidth=0.01)
plot(t, y2, "--", label="y2", color="green")
title("Mixed signal (entire)")
legend()

show()

plot(t, y1, "--", label="y1", color="blue")
plot(t, y, label="y", color="red")
plot(t, y2, "--", label="y2", color="green")

xlim( (0,1) )
title("Mixed signal (zoomed 0-1 sec)")
legend()
show()

NFFT = len(y)
yfft  = abs(fft.fft(y, n=NFFT))
y1fft = abs(fft.fft(y1, n=NFFT))
x = linspace(0, fs, len(yfft))

subplot(121)
title("Entire spectrum")
plot(x, yfft,  'r', linewidth=2, label="mixed")
plot(x, y1fft, 'b', label="sinewave")
xlim( (0,fs/2) )
legend()

subplot(122)
XMIN = x[argmax(yfft)] - fs*0.01
XMAX = x[argmax(yfft)] + fs*0.01
title("Zoomed at peak +- 1%%" % XMAX)
plot(x, yfft, 'r', linewidth=2, label="mixed")
plot(x, y1fft,'b', label="sinewave")
xlim( (XMIN,XMAX) )
legend()

show()
```

## Define utility to plot our tests

```python
hertzfft = None
def plot_mel(signal, fs, mel_filterbank, mel_fft, NFFT):
    
    global hertzfft
    
    print "signal length:", len(signal), "-- fs:", int(fs)
    print "mel filterbank shape:", mel_filterbank.shape
    print "mel fft shape:", mel_fft.shape
    
    title("spectrogram")
    data, freqs, bins, im  = specgram(signal, Fs=fs/1000.)
    xlabel (u"msec")
    ylabel ("kHz")
    autoscale(tight=True)
    show()

    title ("Mel spectrum")
    pcolormesh(mel_filterbank.T)
    autoscale(tight=True)
    show()
    
    print "len Mel spectr:", mel_filterbank.T.shape

    title ("FFT of mel spectrum (matrix)")
    
    y = arange(mel_fft.shape[0])
    x = linspace(0, float(fs)/NFFT, mel_fft.shape[1])
    
    pcolormesh(x, y, mel_fft)
    #xlim( (0,NFFT/2) )
    show()
    
    pcolormesh(mel_fft)
    #xlim( (0,NFFT/2) )
    show()
    
    mel_fs = len(y)/len(mel_feat)
    xmelfft = linspace(0, mel_fs, NFFT)

    print "mel sampling rate:", mel_fs
    
    title ("log10 FFT of mel spectrum (matrix)")
    pcolormesh( 10*log10(mel_fft) )
    xlim( (0,NFFT/2) )
    show()  
    
    suptitle ("FFT of mel spectrum (line plot)")

    subplot(121)
    title ( "full size" )
    lines = plot(xmelfft, mel_fft.T)
    xlim( (0,mel_fs/2) )

    subplot(122)
    title ( "excluding first bin" )
    lines = plot(mel_fft.T)
    #ylim ( (0,10) )
    xlim ( (0,mel_fs/10.) )
    show()
    print len(lines), "lines"
    
```
