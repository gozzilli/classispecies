## Just as a recap, plot a sinewave and its FFT

```python
fs   = 8000.
secs = 5.
freq = 440.

x = linspace(0, secs, secs*fs)
x2 = arange(0, secs, 1./fs)

y = sin(2 * pi * freq * x)

fig = figsize(10,4)
suptitle("%d seconds of a %d Hz sinewave, sampled at %d Hz" % (secs, freq, fs) )
subplot(121)
plot(x, y)

subplot(122)
title("cut after 200 samples")
plot(x[:200], y[:200])

show()

yfft = abs(np.fft.fft(y))
plot(linspace(0, fs, len(yfft)), yfft)
title ("FFT of the sinewave")
xlim((0,fs/2.))
show()
```

## Prepare the data

```python
fs   = 8000.
secs = 10.
freq = 1.

t  = linspace(0, secs, secs*fs)

y1 = sin(2 * pi * freq * t)
y2 = 0.2*sin(4 * pi * freq * t)
y2 = [0] * int(secs*fs)
sl = arange(0, fs*secs, fs*secs/4.).astype(int)

y  = hstack( (y1[sl[0]:sl[1]], y2[sl[1]:sl[2]], y1[sl[2]:sl[3]], y2[sl[3]:None]) )

plot(t, y1, "--")
plot(t, y2, "--")
plot(t, y, linewidth=2)
title("Mixed signal")
show()

yfft = abs(fft.fft(y))
x = linspace(0, fs, len(yfft))

subplot(121)
title("Entire spectrum")
plot(x, yfft, 'r', linewidth=2, label="mixed")
plot(x, abs(fft.fft(y1)), label="sinewave")
xlim( (0,fs/2) )
legend()

subplot(122)
title("Zoomed in 0-5 Hz")
plot(x, yfft, 'r', linewidth=2, label="mixed")
plot(x, abs(fft.fft(y1)), label="sinewave")
xlim( (0,5) )
legend()

show()
```

### Extract mel filterbanks

```python
from features import logfbank, fbank
# alternative options: winlen=0.0232, winstep=0.0232, nfilt=40)

mel_feat = np.clip(np.nan_to_num(logfbank(y, nfilt=40)), 
                       a_min=-100, a_max=100)

title("wave")
plot(t, y)
show()

title("Hertz spectrogram")
_ = specgram(y, Fs=fs/1000)
xlabel (u"μsec")
ylabel ("kHz")
autoscale(tight=True)
show()

title("Mel Log spectrum")
_ = pcolormesh(mel_feat.T)
autoscale(tight=True)
show()
```

## Define utility to plot our tests

```python
def plot_test(signal, fs, mel_filterbank, mel_fft):
    
    print "signal length:", len(signal), "-- fs:", int(fs)
    print "mel filterbank shape:", mel_filterbank.shape
    print "mel fft shape:", mel_fft.shape
    
    title("spectrogram")
    _ = specgram(signal, Fs=fs/1000.)
    xlabel (u"μsec")
    ylabel ("kHz")
    autoscale(tight=True)
    show()

    title ("Mel spectrum")
    pcolormesh(mel_filterbank.T)
    show()

    title ("FFT of mel spectrum (matrix)")
    pcolormesh(mel_fft.T)
    show()

    suptitle ("FFT of mel spectrum (line plot)")

    subplot(121)
    title ( "full size" )
    lines = plot(mel_fft)

    subplot(122)
    title ( "excluding first bin" )
    lines = plot(mel_fft)
    ylim ( (0,10) )
    show()
    print len(lines), "lines"
```

## Now take the FFT over time

```python
melfft = abs(scipy.fft(mel_feat))

plot_test(y, fs, mel_feat, melfft)
```

Just checking that it's acting on a per-bin basis

```python
melfft = []
for melbin in mel_feat.T:
    melfft.append(abs(scipy.fft(melbin)))
    
print len(melfft)
assert len(melfft) == 40

melfft = array(melfft)

print "shape:", melfft.shape, len(melfft)
pcolormesh(melfft.T[0:10])
colorbar()
show()

suptitle ("FFT of mel spectrum (line plot)")

subplot(121)
title ( "full size" )
lines = plot(melfft.T)
xlim ( (0, melfft.shape[1]/2))

subplot(122)
title ( "excluding first bin" )
lines = plot(melfft.T)
#ylim ( (0,1000) )
xlim ( (0, 50) )
show()
print len(lines), "lines"

plot_test(y, fs, mel_feat, melfft)

```

## What would the pure sinewave look like?

```python
signal2 = y1
mel_feat2 = np.clip(np.nan_to_num(logfbank(signal2, nfilt=40)), 
                       a_min=-100, a_max=100)
melfft2 = abs(scipy.fft(mel_feat2))
title ("sinewave")
plot(t, signal2)
show()

title("spectrogram")
_ = specgram(signal2, Fs=fs)
autoscale(tight=True)
show()

title ("Mel spectrum")
pcolormesh(mel_feat2.T)
show()

title ("FFT of mel spectrum (matrix)")
pcolormesh(melfft2)
show()

suptitle ("FFT of mel spectrum (line plot)")

subplot(121)
title ( "full size" )
lines = plot(melfft2)

subplot(122)
title ( "excluding first bin" )
lines = plot(melfft2)
ylim ( (0,10) )
show()
print len(lines), "lines"
```
